# Reading PLINK files 

in this section I used pandas_plink to read bed, bim and fam files. I also used pandas to read the annotation file. I then merged plink files with annotation file as a dataframe. 

```python
import pandas as pd
import pandas_plink as ppl # pandas-plink is a Python package for reading PLINK files into pandas DataFrames.
import os
import numpy as np
import multiprocessing as mp
from tqdm import tqdm
```

```python
# change directory to main folder
os.chdir('../')
```

```python
# Specify the PLINK file paths and unified name (without the extensions of files)
plink_files = 'Data/Input/v54.1'

# Load the PLINK files
(bim, fam, bed) = ppl.read_plink(plink_files)
```

```python
sample_id = fam['iid'].to_list() # saving sample ids in a list
# bim file structure: chr snp cm bp a0(minor Allele) a1(major Allel)
snp_list = bim['snp'].tolist() # we can access the snp names from the bim file
```

```python
# make a dataframe from the genotype data
genotype = pd.DataFrame(bed.compute().T)
genotype.columns = snp_list
genotype.insert(0, "sample_id", sample_id, True)

# make memory free 
bed=0
bim=0
fam=0

genotype
```

```python
# openning annotation file and making a dataframe from it
anno = pd.read_csv('Data/v54.1.anno', sep='\t' , low_memory=False)
anno = pd.DataFrame(anno)
anno.rename(columns={'Date mean in BP in years before 1950 CE [OxCal mu for a direct radiocarbon date, and average of range for a contextual date]': 'date'}, inplace=True)
anno = anno[['Genetic ID', 'date', 'Political Entity','ASSESSMENT','Group ID','Lat.','Long.' ,'Molecular Sex']]
anno.rename(columns={'Genetic ID': 'sample_id', 'Political Entity':'country'}, inplace=True)
anno
```

```python
# merge the genotype and anno dataframes based on the sample_id
genotype = pd.merge(genotype, anno, on='sample_id', how='left')
genotype
```

# Filtering out missing data

I used multiple-step filtering of sequential sample-column filtering out missing data to reach a final dataframe.

```python
genotype = genotype[genotype.isnull().sum(axis=1)/genotype.shape[1] <0.7] # remove samples with more than 70% missing data
genotype= genotype.loc[:, (genotype.isnull().sum(axis=0)/genotype.shape[0]) < 0.2] # remove snps with more than 20% missing data
genotype = genotype[genotype.isnull().sum(axis=1)/genotype.shape[1] <0.1] # remove samples with more than 10% missing data
genotype= genotype.loc[:, (genotype.isnull().sum(axis=0)/genotype.shape[0]) < 0.06] # remove snps with more than 6% missing data
genotype
```

```python
# measure missingness per SNP columns
list_missingness = genotype.isnull().sum(axis=0)/genotype.shape[0]
# plot the missingness per SNP
list_missingness.plot.hist(bins=100, figsize=(10,5), title='Missingness per SNP')
```

```python
genotype.date.plot.hist(bins=100, figsize=(10,5), title='Date distribution')
```

```python
genotype = genotype[
    genotype["date"] < 15000
].copy()  # Filter the data to only include dates before 15000
genotype.loc[:, "date_catg"] = (
    genotype.loc[:, "date"] // 1000
) + 1  # Create a new variable which is the date divided by 1000 and rounded up
genotype.loc[:, "country_date"] = (
    genotype["country"].copy() + "_" + genotype["date_catg"].astype(str).copy()
)  # Create a new variable which combines the country with the date category
genotype
```

Here, I chosed to remove country_date groups with less than 2 samples. 

```python
group=genotype.groupby('country_date').count().sort_values(by=['sample_id'], ascending=False)
# country-date with less than 2 samples
catg_to_remove= group[group['sample_id']<2].index # first one 10
len(catg_to_remove) #110 country-date with less than 2 samples to remove
genotype = genotype[~genotype['country_date'].isin(catg_to_remove)] # remove country-date with less than 2 samples
genotype
```

```python
# bring non SNP columns to the front
cols_to_move = ['sample_id','country', 'date','country_date','ASSESSMENT', 'Group ID', 'Lat.', 'Long.', 'Molecular Sex', 'date_catg']
genotype = genotype[cols_to_move + [col for col in genotype.columns if col not in cols_to_move]]
genotype
```

## Calculating MAF for each SNP inside each country_date group

```python
columns_to_remove = ['sample_id', 'country', 'ASSESSMENT','date', 'Group ID', 'Lat.', 'Long.', 'Molecular Sex', 'date_catg']
genotype = genotype.drop(columns=columns_to_remove)
```

```python
def get_freq(df):
    """
    This function takes a dataframe and returns a list of allele frequencies
    calculated from the genotype data in the dataframe.
    in bed file 0 is homozygous for the minor allele, 1 is heterozygous, and 2 is homozygous for the major allele.
    """
    list_AF = []
    for col in df.columns:
        # fill na with the most frequent value in the column, or 2 if there is no mode
        if df[col].notnull().any():
            mode_value = df[col].mode().iloc[0]
        else:
            mode_value = 2
        df[col].fillna(mode_value, inplace=True)
        # count 0s, 1s, 2s, and missing values
        counts = df[col].value_counts(dropna=False)
        # calculate the frequency of 0s, 1s, 2s, and missing values
        freq = counts / sum(counts)
        # calculate the frequency of 0s*2 + 1s*1
        homo = 0
        hetro = 0
        if 0 in freq.index:
            homo = freq[0]
        if 1 in freq.index:
            hetro = freq[1]
        AF = homo + hetro / 2
        # add column to df_unified
        list_AF.append(AF)

    return list_AF
```

Below I calculated MAF for each SNP inside each country_date group 

The script will filter samples based on country_date groups and then uses function get_maf to calculate MAF for each SNP inside each country_date group.


```python

def process_date(date):
    df = genotype[genotype["country_date"] == date].copy()
    result_AF = get_freq(df)
    result_row = [date] + result_AF[1:]
    return result_row


if __name__ == "__main__":
    col_names = genotype.columns.tolist()
    df_country_date_AF = pd.DataFrame(columns=col_names)

    unique_dates = genotype.country_date.unique()

    # Create a multiprocessing pool
    pool = mp.Pool(15)

    # Process the unique dates in parallel and collect the results
    results = list(
        tqdm(
            pool.imap(process_date, unique_dates),
            total=len(unique_dates),
            desc="Processing",
        )
    )

    # Populate the DataFrame with the results
    for result_row in results:
        df_country_date_AF.loc[len(df_country_date_AF)] = result_row

    # Close the multiprocessing pool
    pool.close()
    pool.join()
```

# Save Results

```python
df_country_date_AF.to_csv('Data/Output/0_country_date_MAF.csv',index=False)
```
